```python
from itertools import combinations
import math
```

```python
def min_coins(value, coins):
    dp = [math.inf] * (value + 1)
    dp[0] = 0
    for v in range(1, value + 1):
        for c in coins:
            if c <= v:
                dp[v] = min(dp[v], dp[v - c] + 1)
    return dp[value]
```

```python
def avg_coins_for_system(coins):
    total = 0
    for x in range(1, 100):
        total += min_coins(x, coins)
    return total / 99
```

```python
best_system = None
best_avg = float('inf')
```

```python
for a, b in combinations(range(2, 100), 2):
    coins = [1, a, b]
    avg = avg_coins_for_system(coins)
    if avg < 5.4 and avg < best_avg:
        best_avg = avg
        best_system = coins
```

```python
print(f"Best system (average < 5.4): {best_system}, Average coins: {best_avg:.3f}")
```
